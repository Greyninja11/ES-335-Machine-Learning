When matrix multiplication is done row by column, for a matrix A(n,k).B(k,m), we see that each k elements of n rows in A get multiplied to each element in m column of B. Therefore the number of multiplications would be  = n.m.k
Similarly the every entry would be scalar product of two k vectors which would entail k-1 additions and since there would be n.m entries the number of additions would be n.m.(k-1).

````python
#Q8            
class Student:
    def __init__(self, roll_num, name):
        self.roll_num = roll_num
        self.name = name
        self.marks = {}

    def addsubjectmarks(self, subject, marks):
        self.marks[subject] = marks


class Branch:
    def __init__(self):
        self.students = []

    def enroll_student(self, student):
        self.students.append(student)


class Year:
    def __init__(self, year):
        self.year = year
        self.branches = {}

    def include_branch(self, branch_name):
        if branch_name not in self.branches:
            self.branches[branch_name] = Branch()

    def add_student_to_branch(self, branch_name, student):
        if branch_name in self.branches:
            self.branches[branch_name].enroll_student(student)
        else:
            self.include_branch(branch_name)
            self.branches[branch_name].enroll_student(student)
````

````python
#Q9
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.5, 100.5, 0.5)
y1 = x
y2 = x ** 2
y3 = x ** 3 / 100
y4 = np.sin(x)
y5 = np.sin(x) / x
y6 = np.log(x)
y7 = np.exp(x)

fig, axs = plt.subplots(3, 3, figsize=(12, 12))

# Plotting each function in a separate subplot
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('y1 = x')

axs[0, 1].plot(x, y2)
axs[0, 1].set_title('y2 = x^2')

axs[0, 2].plot(x, y3)
axs[0, 2].set_title('y3 = x^3 / 100')

axs[1, 0].plot(x, y4)
axs[1, 0].set_title('y4 = sin(x)')

axs[1, 1].plot(x, y5)
axs[1, 1].set_title('y5 = sin(x) / x')

axs[1, 2].plot(x, y6)
axs[1, 2].set_title('y6 = log(x)')

axs[2, 0].plot(x, y7)
axs[2, 0].set_title('y7 = exp(x)')

# Hide the empty subplot
axs[2, 1].axis('off')
axs[2, 2].axis('off')

plt.tight_layout()

plt.show()
````

Broadcasting in numpy is used on arrays during arithmetic operations. Subject to certain constraints, the smaller array is “broadcast” across the larger array so that they have compatible shapes. Broadcasting provides a means of vectorizing array operations so that looping occurs in C instead of Python. This helps improving efficiency and execution time. 
For numpy operation on arrays to be carried out successfully, the two arrays must have the shape. Broadcasting ensures that this occurs. 
Ex: If we are performing any arithmetic operation on an array and a scalar, broadcasting comes in and broadcasts this scalar to the size of the array for the operation.

```
a = np.array([0.0, 10.0, 20.0, 30.0])
``` 
